# importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Loading dataset

```python
df = pd.read_csv("Airline_Delay_Cause.csv")
df.head()
```

# UNIVARIATE ANALYSIS

```python
df["arr_delay"].describe()
```

Histogram (Distribution of delays)

```python
plt.hist(df["arr_delay"], bins=30)
plt.title("Distribution of Arrival Delay")
plt.xlabel("Arrival Delay (minutes)")
plt.ylabel("Frequency")
plt.show()
```

Count of flights per month

```python
df["month"].value_counts().sort_index().plot(kind="bar")
plt.title("Flight Count per Month")
plt.xlabel("Month")
plt.ylabel("Number of Flights")
plt.show()
```

Total delay time for each cause

```python
cause_cols = ["carrier_delay", "weather_delay", "nas_delay", "security_delay", "late_aircraft_delay"]

df[cause_cols].sum().plot(kind="bar")
plt.title("Total Delay by Each Cause")
plt.xlabel("Delay Cause")
plt.ylabel("Total Delay (minutes)")
plt.show()
```

# **BIVARIATE ANALYSIS**
Month vs Average Arrival Delay

```python
df.groupby("month")["arr_delay"].mean().plot(kind="line")
plt.title("Average Arrival Delay by Month")
plt.xlabel("Month")
plt.ylabel("Avg Delay (minutes)")
plt.show()
```

Carrier Delay vs Arrival Delay

```python
plt.scatter(df["carrier_delay"], df["arr_delay"])
plt.title("Carrier Delay vs Arrival Delay")
plt.xlabel("Carrier Delay")
plt.ylabel("Arrival Delay")
plt.show()
```

Weather Delay vs Arrival Delay

```python
plt.scatter(df["weather_delay"], df["arr_delay"])
plt.title("Weather Delay vs Arrival Delay")
plt.xlabel("Weather Delay")
plt.ylabel("Arrival Delay")
plt.show()
```

NAS Delay vs Arrival Delay

```python
plt.scatter(df["nas_delay"], df["arr_delay"])
plt.title("NAS Delay vs Arrival Delay")
plt.xlabel("NAS Delay")
plt.ylabel("Arrival Delay")
plt.show()
```

I performed univariate and bivariate analysis on the Airline Delay Cause dataset.

In univariate analysis, I studied each column separately. I plotted the distribution of arrival delay using a histogram. I also checked how many flights occurred in each month and which delay cause had the highest total delay.

In bivariate analysis, I compared two columns at a time. I plotted how the average arrival delay changes each month. Then I used scatter plots to see the relationship between arrival delay and carrier delay, weather delay, and NAS delay. This helped me understand which delay cause affects arrival delay more.

The analysis showed clear patterns and gave me an understanding of airline delays.
